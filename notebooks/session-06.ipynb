```python
df = df_copy.copy()


def fix_own_occupied(df):
    df = df.copy()
    for i, row in enumerate(df["OWN_OCCUPIED"]):
        if row in ["Y", "N"]:
            continue
        df.loc[i, "OWN_OCCUPIED"] = np.nan
    return df


display("Before function")
display(df)
df_cleaned = fix_own_occupied(df)
display("After function")
display(df_cleaned)
```

```python
fix_own_occupied(df)
```

```python
def fix_own_occupied_numpy(df):
    df = df.copy()
    df["OWN_OCCUPIED"] = np.where(
        df["OWN_OCCUPIED"].isin(["Y", "N"]), df["OWN_OCCUPIED"], np.nan
    )
    return df
```
