```python
india = df_can.loc["India", years]
india.index = india.index.map(int)

india.plot()

plt.title("Immigration from India")
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")

plt.text(x=1999, y=16000, s="Text di sini")

plt.show()  # kaya print
```

```python
china_india = df_can.loc[["China", "India"], years]
china_india.head()
```

```python
df_china_india = china_india.transpose()
df_china_india.head()
```

```python
df_china_india.plot()

plt.title("Immigrants from China and India")
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")

plt.show()
```

```python
top_origin = df_can[years].sum(axis=1).sort_values(ascending=False)[:5].index.to_list()
top_origin
```

```python
top_5 = df_can.loc[top_origin, years]
top_5.transpose().plot()
```

```python
top_5.T.loc[years[10:20]].plot(kind="barh")
```

```python
fig = plt.figure()

ax0 = fig.add_subplot(2, 2, 1)
ax1 = fig.add_subplot(2, 2, 2)

top_5.T.loc[years[10:20]].plot(kind="area", figsize=(20, 5), ax=ax0)
ax0.legend(loc="upper left")

china_india.T.loc[years[10:20]].plot(kind="line", figsize=(20, 5), ax=ax1)
ax0.legend(loc="upper left")

plt.show()
```
